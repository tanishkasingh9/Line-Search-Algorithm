## Backtracking Line Search Algorithm

### Define F(x), F'(x) and step size

```
def funct(x):
  f_x = round(x[0]**2+5*x[1]**2,5)
  return f_x
def del_func(x):
  f_partial = [-2*x[0],-10*x[1]]
  return f_partial
def step_size(t):
  t = round(t+0.001,5)
  return t

```

### Exact Line Search

```
def exact_line_search(x,t):
  f = []
  x_r_1 = []
  x_r_2 = []
  f_x = funct(x)
  f.append(f_x)
  x_r_1.append(x[0])
  x_r_2.append(x[1])
  min=[f_x,t]
  for i in range(1,500):
    print('iteration ',i,'stepsize = ',t, min)
    d = del_func(x)
    x_new = [round(x[0]+t*d[0],4),round(x[1]+t*d[1],4)]
    x_r_1.append(x_new[0])
    x_r_2.append(x_new[1])
    f_xnew = funct(x_new)
    f.append(f_xnew)
    x = x_new
    f_x = f_xnew
    if f_xnew < min[0]:
      min[0] = f_xnew
      min[1] = t
    else:
      pass
    t = step_size(t)
  return f,x_r_1,x_r_2,min[1]
```

### Running Exact Line Search for finding step size starting with step size 1

```
f,x_r_1,x_r_2,t = exact_line_search([5,1],0.05)
```

### Plot step size vs iterations

```
x_1 = np.arange(-5, 5)
x_2 = np.arange(-5, 5)
z = np.zeros((10, 10))
for i in range(0,len(x_1)):
  for j in range(0,len(x_2)):
    z[i][j] = x_1[i]**2+5*x_2[j]**2
fig = plt.figure(figsize=(10,10))    
cs = plt.contour(x_1,x_2,z, levels=[i for i in range(-70,500,5)])
plt.plot(x_r_1,x_r_2)
plt.scatter(x_r_1,x_r_2)


```
